```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('owid-covid-data.csv')
```

```python
df.drop(['new_deaths_smoothed','new_cases_per_million','total_cases_per_million'],axis=1,inplace=True)
```

```python
df
```

```python
df.rename(columns={'date': 'Date','location':'Country','continent': 'Continent','iso_code':'ISO_code'},inplace=True)
```

```python
df
```

```python
continent_unique = list(df.Continent.unique())
```

```python
continent_unique
```

```python
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

```python
df3 = df2.groupby(['Date','Country',])[['Date','Country','total_cases','total_deaths','total_vaccinations']].sum().reset_index()
```

```python
df3['total_cases'].replace({'missing_value': 0},inplace=True)
```

```python
df3['total_deaths'].replace({'missing_value': 0},inplace=True)
```

```python
df3['total_vaccinations'].replace({'missing_value': 0},inplace=True)
```

```python
df4 = df3[df3['total_deaths']>100000]
```

```python
countries = df4['Country'].unique()
```

```python
len(countries)
```

```python
country_deaths_greaterthan1000000 = list(df4.Country.unique())
```

```python
country_deaths_greaterthan1000000
```

```python
for idx in range(0,len(countries)):
    c = df4[df4['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(c)),c['total_cases'],color="blue",lable="total_cases")
    plt.scatter(np.arange(0,len(c)),c['total_deaths'],color="red",lable="total_deaths")
    plt.scatter(np.arange(0,len(c)),c['total_vaccinations'],color="green",lable="total_vaccinations")
    plt.title(countries[idx])
    plt.xlable("number of days since first suspect")
    plt.ylable("number of cases")
    plt.legend
    plt.show()
```
